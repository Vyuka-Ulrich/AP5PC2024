### Zadání: Simulace jednoduchých elektrických komponent

Vaším úkolem je vytvořit jednoduchý program v jazyce c++, který bude simulovat základní vlastnosti elektrických komponent, jako jsou rezistor a žárovka. Program by měl využívat abstraktní třídu pro základní popis komponent a od ní odvozené konkrétní třídy.

---

### **Požadavky na implementaci**

1. **Abstraktní třída `Component`**
   - Vytvořte abstraktní třídu `Component`, která bude obsahovat:
     - Metodu `print`, která vypíše informace o komponentě.
     - Metodu `compute_power_consumption(voltage)`, která vypočítá příkon při zadaném napětí.

2. **Konkrétní třídy komponent**
   - **Rezistor (`Resistor`)**
     - Třída musí obsahovat atribut odporu (v ohmech).
     - Metoda `compute_power_consumption` spočítá příkon podle vzorce:
       \[
       P = \frac{U^2}{R}
       \]
     - Metoda `print` vypíše: „Resistor: Resistance = `<hodnota>` ohms“.
   - **Žárovka (`Bulb`)**
     - Třída musí obsahovat atributy příkonu (ve wattech) a napětí (ve voltech).
     - Metoda `compute_power_consumption` vždy vrací příkon žárovky, protože je na zadané napětí navržena.
     - Metoda `print` vypíše: „Bulb: Power = `<hodnota>` W, Voltage = `<hodnota>` V“.

3. **Hlavní funkce**
   - V hlavní funkci vytvořte seznam komponent (`list`).
   - Přidejte do seznamu následující komponenty:
     - Rezistor s odporem 100 ohmů.
     - Žárovka s příkonem 60 W a napětím 230 V.
   - Pro každou komponentu:
     - Vypište informace o komponentě (`print`).
     - Vypište její příkon při napětí 230 V (`compute_power_consumption`).

---
