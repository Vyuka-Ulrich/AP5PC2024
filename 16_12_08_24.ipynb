
# Zadání: Geometrické tvary a potřeba barvy

## Cíl
Vytvořte program, který modeluje geometrické tvary, vypočítá jejich plochy a odhaduje množství barvy potřebné k jejich pokrytí.

---

### 1. Abstraktní třída `Shape`
Vytvořte abstraktní třídu `Shape`, která obsahuje:
- **Čistě virtuální metody**:
  - `void print() const`: Vypíše vlastnosti tvaru (např. délky, poloměr atd.).
  - `T computeArea() const`: Vypočítá plochu tvaru.
- **Implementovanou metodu**:
  - `T estimatePaintRequired(T paintCoverage) const`: Odhadne množství barvy potřebné k pokrytí tvaru na základě jeho plochy a krytí barvy (`paintCoverage`).
    - `paintCoverage` označuje, kolik čtverečních jednotek může být pokryto jednou jednotkou barvy.
    - Výpočet:
      ```
      Potřebná barva = Plocha / Krytí barvy
      ```

---

### 2. Třída `Circle`
Vytvořte třídu `Circle`, která dědí z `Shape` a má:
- **Atributy**:
  - `T radius`: Poloměr kruhu.
- **Implementaci metod**:
  - `print`: Vypíše poloměr kruhu.
  - `computeArea`: Vypočítá plochu kruhu pomocí vzorce:
    ```
    Plocha kruhu = π * radius^2
    ```

---

### 3. Třída `Rectangle`
Vytvořte třídu `Rectangle`, která dědí z `Shape` a má:
- **Atributy**:
  - `T length`: Délka obdélníku.
  - `T width`: Šířka obdélníku.
- **Implementaci metod**:
  - `print`: Vypíše délku a šířku obdélníku.
  - `computeArea`: Vypočítá plochu obdélníku pomocí vzorce:
    ```
    Plocha obdélníku = length * width
    ```

---

### 4. Hlavní program
1. Vytvořte seznam (`std::vector<std::shared_ptr<Shape<T>>>`), který obsahuje různé tvary (několik kruhů a obdélníků).
2. Pomocí metody `print` vypište vlastnosti každého tvaru.
3. Pro každý tvar:
   - Spočítejte jeho plochu pomocí `computeArea`.
   - Odhadněte množství barvy potřebné k pokrytí tvaru při `paintCoverage = 1.0`.
4. Seřaďte tvary podle jejich plochy v rostoucím pořadí (pomocí `std::sort`).
5. Vypište seřazený seznam tvarů.
