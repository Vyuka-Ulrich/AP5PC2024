
# Zadání: Simulace fyzikálních těles

## Cíl
Vytvořte program, který modeluje fyzikální tělesa a jejich vlastnosti, včetně výpočtu hmotnosti a působící síly.

---

### 1. Abstraktní třída `PhysicalBody`
Vytvořte abstraktní třídu `PhysicalBody`, která obsahuje:
- **Čistě virtuální metody**:
  - `void print() const`: Vypíše vlastnosti tělesa (např. hmotnost, hustotu atd.).
  - `T computeMass() const`: Vypočítá hmotnost tělesa.
- **Implementovanou metodu**:
  - `T computeForce(T acceleration) const`: Vypočítá sílu působící na těleso při zadaném zrychlení \( a \) pomocí vzorce:
    ```
    F = m * a
    ```
    kde \( F \) je síla, \( m \) je hmotnost tělesa a \( a \) je zrychlení.

---

### 2. Třída `Sphere`
Vytvořte třídu `Sphere`, která dědí z `PhysicalBody` a má:
- **Atributy**:
  - `T radius`: Poloměr koule.
  - `T density`: Hustota materiálu koule.
- **Implementaci metod**:
  - `print`: Vypíše poloměr a hustotu koule.
  - `computeMass`: Vypočítá hmotnost koule na základě jejího objemu a hustoty:
    ```
    Objem koule = (4/3) * π * radius^3
    Hmotnost koule = Objem koule * hustota
    ```

---

### 3. Třída `Cuboid`
Vytvořte třídu `Cuboid`, která dědí z `PhysicalBody` a má:
- **Atributy**:
  - `T length`: Délka kvádru.
  - `T width`: Šířka kvádru.
  - `T height`: Výška kvádru.
  - `T density`: Hustota materiálu kvádru.
- **Implementaci metod**:
  - `print`: Vypíše rozměry a hustotu kvádru.
  - `computeMass`: Vypočítá hmotnost kvádru na základě jeho objemu a hustoty:
    ```
    Objem kvádru = length * width * height
    Hmotnost kvádru = Objem kvádru * hustota
    ```

---

### 4. Hlavní program
1. Vytvořte seznam (`std::vector<std::shared_ptr<PhysicalBody<T>>>`), který obsahuje různé tělesa (několik koulí a kvádrů).
2. Pomocí metody `print` vypište vlastnosti každého tělesa.
3. Pro každé těleso:
   - Spočítejte jeho hmotnost pomocí `computeMass`.
   - Spočítejte sílu působící na těleso při zrychlení \( a = 9.81 \, 	ext{m/s}^2 \).
4. Seřaďte tělesa podle jejich hmotnosti v rostoucím pořadí (pomocí `std::sort`).
5. Vypište seřazený seznam těles.
